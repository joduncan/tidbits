```python
%matplotlib inline

import itertools
import subprocess
import matplotlib
import numpy as np
import matplotlib.pyplot as plt

def getSamplesFromFiles(files):
    tempSamples = []
    for sampleDoc in files:
        with open(sampleDoc) as fp:
            # do some stuff
            tempSamples += [int(line.rstrip()) for line in fp]
    return tempSamples

def getSamplesFromDGenerator(argsArray):
    tempSamples = []
    output = ""
    for sampleRunArgs in argsArray:
        cmdFlags = ["--" + arg for arg in sampleRunArgs.keys()]
        # this is a little fugly.
        cmdArgs = ["./sample_gen"] + list(itertools.chain.from_iterable(zip(cmdFlags, [str(x) for x in sampleRunArgs.values()])))
        output = subprocess.check_output(cmdArgs)
        outputArray = output.split('\n')
        tempSamples += [int(line.rstrip()) for line in outputArray if len(line) > 0]
    return tempSamples

def generateSamples(argsArray):
    tempSamples = []
    for args in argsArray:
        baseline = args["offset"]-args["range"] if (args["offset"]-args["range"] > 0) else 0
        for sample in range(0, args["samples"]):
            tempSamples.append(int(np.round(baseline + np.random.rand()*args["range"])))
        # TODO: save samples to files.
    return tempSamples

# 1. read files with samples into big array
# 2. plot values, collated averages, etc.
# 3. ???
# 4. profit!

generatorParams = [
#              {"samples": 150, "offset": 60, "range": 20}
              {"samples": 20000, "offset": 30, "range": 30},
              {"samples": 15000, "offset": 70, "range": 40},
              {"samples": 150, "offset": 100, "range": 10},
              {"samples": 5000, "offset": 80, "range": 20},
              {"samples": 10000, "offset": 20, "range": 50}
              ]
#sampleFiles = ["samples50_offset50_range20.txt", "samples50_offset75_range25.txt"]
#sampleFiles=  ["samples2000000_offset50_range30.txt"]
#samples = getSamplesFromFiles(sampleFiles)
#samples2 = getSamplesFromDGenerator(generatorParams)

samples = generateSamples(generatorParams)

print len(samples), np.mean(samples)
#print len(samples2), np.mean(samples2)
#print len(samples+samples2), np.mean(samples+samples2)
```

```python
historic_average_sample_rates = {
                                 "x-small": 2*10,        # 10 sec, this is our detection window.
                                 "small": 2*60*60,       # 1 hour
                                 "medium": 2*60*120,     # 2 hours
                                 "large": 2*60*180,      # 3 hours
                                 "x-large": 2*60*360,    # 6 hours
                                 "xx-large": 2*60*720,   # 12 hours
                                 "xxx-large": 2*60*1440  # 24 hours
                                 }
print historic_average_sample_rates
historic_average_legends = {
                            "x-small": "purple",
                            "small": "darkred",
                            "medium": "red",
                            "large": "orange",
                            "x-large": "yellow",
                            "xx-large": "green",
                            "xxx-large": "lightblue"
                            }
working_set = {
               "x-small": [],
               "small": [],
               "medium": [],
               "large": [],
               "x-large": [],
               "xx-large": [],
               "xxx-large": []
               }
historic_averages = {
                     "x-small": [],
                     "small": [],
                     "medium": [],
                     "large": [],
                     "x-large": [],
                     "xx-large": [],
                     "xxx-large": []
                     }
historic_average = {
                     "x-small": 0,
                     "small": 0,
                     "medium": 0,
                     "large": 0,
                     "x-large": 0,
                     "xx-large": 0,
                     "xxx-large": 0
                     }

samples_watched = 0

# FIXME: averages are no good, because if they're too short, it'll be hard to detect a desired condition vs. typical
# "noise."
# next steps/thoughts: consider outliers and quartiles, where desired conditions have tight quartiles and little
# to no outliers in the bottom 25%. typical noise should have a pretty wide "middle-50%" band, and a healthily
# high % in the bottom quartile.
for sample in samples:
    samples_watched+=1
    for ha in historic_average_sample_rates.keys():
        avg_factor = historic_average_sample_rates[ha]
        if samples_watched < historic_average_sample_rates[ha]:
            avg_factor = samples_watched
        if samples_watched > historic_average_sample_rates[ha]:
            working_set[ha].pop()

        current_average = int(np.round(float(sum(working_set[ha]) + sample)/avg_factor))
#        if(ha == "medium"):
#            print "sample: ", sample, " ", ha, " average: ", current_average
        historic_averages[ha].insert(0, current_average)
        working_set[ha].insert(0, sample)

for i in historic_averages.keys():
    print i, " averages calculated: ", len(historic_averages[i])
```

```python
# now to do an example plot of the data points.

indices = range(0,len(samples))
plt.figure(num=1, figsize=(12, 6))

plt.scatter(indices, samples, c='black', label="samples", alpha=0.6, marker=".", edgecolors="face")

plt.legend()

plt.show()
```
